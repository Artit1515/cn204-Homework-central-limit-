```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Advertising.csv', index_col=0).head(20)
data[["TV","sales"]]
```

```python
plt.scatter(data['TV'], data['sales'])
plt.xlabel('TV  Budget')
plt.ylabel('Sales')
plt.title('Sales and TV linear relation ')
plt.show()
```

```python
x = data["TV"].values
y = data["sales"].values

x_mean = x.mean()
y_mean = y.mean()


xy = np.sum((x - x_mean) * (y - y_mean))
xx = np.sum((x - x_mean)**2)

b1 = xy / xx
b0 = y_mean - b1 * x_mean

print("Regression line =",b0, "+",b1, "* x")
```
